```python
import numpy as np
import matplotlib.pyplot as plt
from solver import time_step_damped_wave
```

```python
# parameters
L     = 1.0     # domain length
c     = 1.0     # wave speed
alpha = 0.1     # damping coefficient
nx    = 40    # number of spatial subintervals
dt    = 0.005   # time step


# 4 sublplts
figure, axis = plt.subplots(2, 2, figsize=(10, 8))

figure.suptitle("Damped Wave with 40 subintervals", fontsize=16)

x1, U_num1, U_ana1 = time_step_damped_wave(L, c, alpha, nx, dt, 1)
axis[0,0].plot(x1, U_num1, 'o-', label="Numerical")
axis[0,0].plot(x1, U_ana1, '-', label="Analytical")
axis[0,0].set_xlabel("x")
axis[0,0].set_ylabel("u(x,T)")
axis[0,0].set_title(f"t=1.00 sec")
axis[0,0].legend()

x2, U_num2, U_ana2 = time_step_damped_wave(L, c, alpha, nx, dt, 4)
axis[0,1].plot(x2, U_num2, 'o-', label="Numerical")
axis[0,1].plot(x2, U_ana2, '-', label="Analytical")
axis[0,1].set_xlabel("x")
axis[0,1].set_ylabel("u(x,T)")
axis[0,1].set_title(f"t=4.00 sec")
axis[0,1].legend()

x3, U_num3, U_ana3 = time_step_damped_wave(L, c, alpha, nx, dt, 9)
axis[1,0].plot(x3, U_num3, 'o-', label="Numerical")
axis[1,0].plot(x3, U_ana3, '-', label="Analytical")
axis[1,0].set_xlabel("x")
axis[1,0].set_ylabel("u(x,T)")
axis[1,0].set_title(f"t=9.00 sec")
axis[1,0].legend()

x4, U_num4, U_ana4 = time_step_damped_wave(L, c, alpha, nx, dt, 16)
axis[1,1].plot(x4, U_num4, 'o-', label="Numerical")
axis[1,1].plot(x4, U_ana4, '-', label="Analytical")
axis[1,1].set_xlabel("x")
axis[1,1].set_ylabel("u(x,T)")
axis[1,1].set_title(f"t=16.00 sec")
axis[1,1].legend()

# set the same x and y limits for all subplots
for ax in axis.flat:
    ax.set_xlim(0, L)
    ax.set_ylim(-1.5, 1.5)
    ax.grid(ls='-.')

plt.tight_layout()
plt.show()
```

```python
# parameters
L     = 1.0     # domain length
c     = 1.0     # wave speed
alpha = 0.5     # damping coefficient
nx    = 40    # number of spatial subintervals
dt    = 0.01   # time step

amplitudes = []
times = []
t = 0
while (t < 10):
    x, U_num, U_ana = time_step_damped_wave(L, c, alpha, nx, dt, t)
    if (max(U_num) > 0):
        amplitudes.append(max(U_num))
    else:
        amplitudes.append(min(U_num))
    times.append(t)
    t += dt

plt.plot(times, amplitudes, 'o-', label="Numerical")
plt.xlabel("Time")
plt.ylabel("Max Displacement")
plt.title("Max Displacement vs Time (" + r'$\alpha$' + " = 0.5)")
plt.grid(ls='-.')
```
